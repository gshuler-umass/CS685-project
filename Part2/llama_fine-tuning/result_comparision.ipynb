```python
def compute_metrics(df, model_output_col, true_label_col, true_value, false_value, model_name):

  true_positives = df[(df[model_output_col] == true_value) & (df[true_label_col] == true_value)].shape[0]
  true_negatives = df[(df[model_output_col] == false_value) & (df[true_label_col] == false_value)].shape[0]
  false_positives = df[(df[model_output_col] == true_value) & (df[true_label_col] == false_value)].shape[0]
  false_negatives = df[(df[model_output_col] == false_value) & (df[true_label_col] == true_value)].shape[0]

  precision = true_positives / (true_positives + false_positives)
  recall = true_positives / (true_positives + false_negatives)
  accuracy = (true_positives + true_negatives) / len(df)
  f1_score = 2 * (precision * recall) / (precision + recall)

  return {
      'model_name' : model_name,
      'precision': precision,
      'recall': recall,
      'accuracy': accuracy,
      'f1_score': f1_score
  }

```

```python

rndm_metrics = compute_metrics(rndm_df, model_output_col = 'baseline_rndm_pred', true_label_col = 'answer', \
                                 true_value ='yes', false_value ='no', model_name= 'rndm')



llama2_metrics = compute_metrics(llama2_df, model_output_col = 'llama2_response_processed', true_label_col = 'answer', \
                                 true_value ='yes', false_value ='no', model_name= 'llama2')
llama3_metrics = compute_metrics(llama3_df, model_output_col = 'llama3_response_processed', true_label_col = 'answer', \
                                 true_value ='yes', false_value ='no', model_name= 'llama3')
llama2_finetuned_metrics = compute_metrics(llama2_finetuned_df, model_output_col = 'llama2_ft_response_processed', true_label_col = 'answer', \
                                           true_value ='yes', false_value ='no', model_name= 'llama2-ft')
llama3_finetuned_metrics = compute_metrics(llama3_finetuned_df, model_output_col = 'llama3_ft_response_processed', true_label_col = 'answer', \
                                           true_value ='yes', false_value ='no', model_name= 'llama3-ft')

print(rndm_metrics)

print(llama2_metrics)
print(llama3_metrics)

print(llama2_finetuned_metrics)
print(llama3_finetuned_metrics)
```
